```python
import psycopg2
import pandas as pd
#connection with postgresql
def connect():
  try:
     conn=psycopg2.connect(database="musicperk", user = "postgres", password = "ASdf@123", host = "localhost", port = "5432")
     cur=conn.cursor()
  except (Exception, psycopg2.DatabaseError) as error:

      print("Error while creating PostgreSQL table", error)
  return conn, cur
#creating table for excel filedata insertion 
def create_table():
    conn,cur=connect()
    cur.execute('CREATE TABLE HINDALCO_1D (datetime TIMESTAMP, close float4,high float4,low float4,open float4,volume float8,instrument varchar(40))')
    conn.commit()
    print("done")
#function for inserting data
def insert_data(datetime,close,high,low,open,volume,instrument):
    conn,cur=connect()
    cur.execute('INSERT INTO  HINDALCO_1D VALUES(%s,%s,%s,%s,%s,%s,%s)',(datetime,close,high,low,open,volume,instrument))
    conn.commit()
connect()
create_table()
```

```python
#loading excel file
data = pd.read_excel ('HINDALCO_1D.xlsx', engine='openpyxl',)
df = pd.DataFrame(data)
datetime=df['datetime']
close=df['close']
high=df['high']
low=df['low']
open=df['open']
volume=df['volume']
instrument=df['instrument']
for i in range(len(close)):
    print(f"item {i} inserted")
    insert_data(datetime[i],float(close[i]),float(high[i]),float(low[i]),float(open[i]),float(volume[i]),str(instrument[i]))
print("done")
```

```python
#calculating simple_moving_average of all column in excel file with n=10
#sma=(c1+c2+c3+......+cn)/n
sma_close=[]
sma_open=[]
sma_volume=[]
sma_high=[]
for j in range(10,len(close)):
    sma_close.append(sum(close[j-10:j])/10)
    sma_open.append(sum(open[j-10:j])/10)
    sma_volume.append(sum(volume[j-10:j])/10)
    sma_high.append(sum(high[j-10:j])/10)
#print(sma_open)
#print(sma_close)
```
